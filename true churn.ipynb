```python
#balancing the dataset(in this data we have more 0 and less 1 so our data is unbalance we have to balnce it)

#undersampling(work for majority data for eg: if we have 10000 data then it will make 1000)
#disadvantage is(it will loss the data)
#oversampling (work for minority data for eg: if we have 1000 data then it will make 10000)
#disadvantage is(it will have duplicate value)
#iblearn(sentatic minority over sampling technique)(making minority class to mejority class)(in this it will remove the both disadvantage)
```

```python
#value counts of class 0 and 1
#(here both are same)
count_class_0,count_class_1 = data.churn_mapped.value_counts()

#divide by class
data_class_0 = data[data['churn_mapped']==0]
data_class_1 = data[data['churn_mapped']==1]
```

```python
data_class_0.shape
```

```python
data_class_1.shape
```

```python
#undersample 0-class and concat the dataframes of both class

a=data_class_0.sample(count_class_1)
data_test_under = pd.concat([a,data_class_1],axis=0)
print("Random under-sampling:")
print(data_test_under.churn_mapped.value_counts())
```

```python
a.shape
```

```python
data_test_under["churn_mapped"].value_counts()
```

```python
x=data_test_under.drop("churn_mapped",axis=1)
x
```

```python
y=data_test_under["churn_mapped"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
#taking the logistic regression algorithm(sklearn.linear_model)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #checking the accuracy (it will be same in all the model)

model = LogisticRegression()

#model fitting(fit keyword is use to train the model)
model.fit(x_train,y_train)

#testing tht model(predict keyword is use to test the model)
pred=model.predict(x_test)

#model accuracy
print(accuracy_score(y_test,pred))
print(pred)
```
